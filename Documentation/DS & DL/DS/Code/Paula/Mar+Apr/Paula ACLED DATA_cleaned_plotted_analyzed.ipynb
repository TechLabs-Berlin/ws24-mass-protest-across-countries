```python
# @title datetime situations
from datetime import datetime
df['event_date'] = pd.to_datetime(df['event_date'])
df['event_date'] = df['event_date'].dt.date
df['event_date'] = pd.to_datetime(df['event_date'])
```

```python
# @title year, month, day, weekday
df['year'] = df['event_date'].dt.year
df['month'] = df['event_date'].dt.month_name() # .str[:3]
df['day'] = df['event_date'].dt.day

df['weekday'] = df['event_date'].dt.day_name().str[:3]
# df['weekday']
```

```python
# @title the beauty of uppercase .....
# df['title'] = df['title'].str.capitalize()
df['title'] = df['title'].str.upper()
# .replace() VERSION
df['title'] = df['title'].replace(
    to_replace=[r'^ '],
    value='',
    regex=True )
'''
df['description'] = df['description'].str.lower()
df['description'] = df['description'].str.capitalize()
# .replace() VERSION
df['description'] = df['description'].replace(
    to_replace=[r'^ '],
    value='',
    regex=True )
    '''
```
